```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(
    r"c:\Users\shuhr\Downloads\DimCustomers.csv",
    sep=",",
    encoding="latin1",
    quotechar='"',
    on_bad_lines="skip"   
)


print(df.head())

```

```python
print(df.isnull().sum())
print(df.duplicated().sum())
```

```python
from datetime import datetime

df['Age'] = pd.to_datetime('today').year - pd.to_datetime(df['BirthDate']).dt.year
```

```python
sns.histplot(df['YearlyIncome'], bins=30, kde=True)
plt.show()
```

```python
sns.boxplot(x='TotalChildren', y='YearlyIncome', data=df)
plt.show()
```

```python
sns.histplot(data=df, x='Age', hue='Gender', bins=20)
plt.show()
```
